### ⚛️ Code Pattern Mechanics

*Unlike previous linear scripts, this project utilizes an Object-Oriented Framework to model progressive time-series data dynamically.*

1.  **The Class `ThermodynamicLabor`:** We encapsulate the Quantum Circuit inside a Python Object. This represents our artificial battery cell.
2.  **`charge_cell()` Method:** Injects maximum energy (Placing the Qubit into a high-energy `|1>` state).
3.  **`experience_state_after_state()` Method:** Iteratively applies amplitude damping (decay). This mimics the battery running an electric motor, laboring heavily, and steadily relinquishing its internal chemical charge.
4.  **`dynamic_readout()`:** Instantly prints terminal-based visual telemetry for every epoch, breaking the rigid framework of relying solely on heavy end-line graphing libraries.

```
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import math

# Setting the physics simulation seed exactly to 84 (Surah Al-Inshiqaq)
np.random.seed(84)

class ThermodynamicLabor:
    
    def __init__(self, damping_factor=0.15):
        # Initializes the structural constraints of the EV Battery Cell.
        self.simulator = AerSimulator()
        self.damping_factor = damping_factor
        self.current_state_energy = 100.0  # Start fully charged
        print("🔋 [BATTERY ASSEMBLED] Quantum Thermodynamic Cell verified.")
        
    def charge_cell(self):
        # Inevitable Creation: Pushing the system into maximum potential energy.
        self.qc = QuantumCircuit(1, 1)
        self.qc.x(0)  # High Energy State |1>
        print("⚡ [POWER SURGE] The cell is holding absolute charge. Ready for exertion.")
        
    def _apply_decay_barrier(self, epoch):
        # Mathematical Amplitude Damping to simulate physical degradation.
        # We simulate the fractional decay of the state based on the epoch friction
        decay_angle = math.acos(math.sqrt(1 - (self.damping_factor * epoch)))
        self.qc.rx(decay_angle, 0)

    def experience_state_after_state(self, total_epochs=5):
        # Executes sequential time-evolution and outputting dynamic telemetry.
        print("\n=======================================================")
        print("   ⚛️ DYNAMIC THERMODYNAMIC EVOLUTION TRACKER ")
        print("=======================================================")
        
        for epoch in range(1, total_epochs + 1):
            # Every epoch, the battery labors and decays further
            self._apply_decay_barrier(epoch)
            
            # Temporary measurement for the local state
            temp_qc = self.qc.copy()
            temp_qc.measure(0, 0)
            
            compiled = transpile(temp_qc, self.simulator)
            job = self.simulator.run(compiled, shots=1024)
            counts = job.result().get_counts()
            
            # Extracting the remaining energy (Probability of staying in |1>)
            energy_held = counts.get('1', 0)
            percentage = (energy_held / 1024) * 100
            
            # Creating an embedded ASCII progress bar
            bar_fill = int(percentage / 5)
            visual_bar = "█" * bar_fill + "-" * (20 - bar_fill)
            
            print(f"Epoch {epoch} | State Transition | Exertion Active")
            print(f"Charge Retained: [{visual_bar}] {percentage:.1f}%")
            print("-------------------------------------------------------")
            
        print("⚠️ [EXHAUSTION] The quantum cell has relinquished its contents fully.")
        print("=======================================================\n")

# Procedural Execution Block
ev_battery = ThermodynamicLabor(damping_factor=0.18)
ev_battery.charge_cell()
ev_battery.experience_state_after_state(total_epochs=5)
```
