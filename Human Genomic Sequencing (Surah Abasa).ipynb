### ⚙️ How Our Code Works

1. **Setup DNA Qubits:** We create a DNA pool represented by 3 Qubits (yielding an 8-sequence genetic fragment).
2. **Biological Initialization:** We randomize all possibilities to represent an un-sequenced gene strand.
3. **Genetic Tagging (Oracle):** The system flags sequence `|111>` as the target 'destined' mutation.
4. **DNA Refinement:** The diffuser cleans the genetic noise, making the target sequence prominent.
5. **Path Eased (Measurement):** We extract our final genetic reading, securing the exact location of the gene.

```
import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile
from qiskit_aer import AerSimulator

# Set seed to 80 (Surah Abasa)
np.random.seed(80)

print("🧬 [PHASE 1] Accessing Genomic Database... Loading Sperm-Drop RNA Sequences...")

dna_qr = QuantumRegister(3, 'dna_strand')
dna_cr = ClassicalRegister(3, 'extraction')
qc = QuantumCircuit(dna_qr, dna_cr)

# Step 1: Initialize the full biological sequence array
qc.h(dna_qr)
qc.barrier()
print("🔬 [PHASE 2] Superposition Active: All 3-Billion Base-Pair iterations live in memory.")

# Step 2: The Oracle (Tagging the 'Destined' Marker |111>)
qc.mcx([dna_qr[0], dna_qr[1]], dna_qr[2])
qc.barrier()
print("🩸 [PHASE 3] Genetic Oracle Triggered: The target anomalous trait has been successfully tagged.")

# Step 3: Amplifying the specific target trait over the generic DNA noise
qc.h(dna_qr)
qc.x(dna_qr)
qc.h(dna_qr[2])
qc.mcx([dna_qr[0], dna_qr[1]], dna_qr[2])
qc.h(dna_qr[2])
qc.x(dna_qr)
qc.h(dna_qr)
qc.barrier()
print("🧪 [PHASE 4] Amplification applied. The path to the specific sequence is eased.")

# Step 4: Extract the sequencing
qc.measure(dna_qr, dna_cr)

simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1024)
counts = job.result().get_counts()

target_sequence = counts.get('111', 0)
print("\n================================================")
print("          🧬 HUMAN GENOMIC SEQUENCING REPORT     ")
print("================================================")
print(f"Target Genomic Marker [111] successfully sequenced and isolated.")
print(f"Confidence Isolation Value: {(target_sequence / 1024) * 100:.2f}% accuracy against chaotic biological noise.")
print("The path has been mapped. Medical protocols can proceed.")
print("================================================")
```
