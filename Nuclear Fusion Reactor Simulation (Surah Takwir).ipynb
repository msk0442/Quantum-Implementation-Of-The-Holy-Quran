### ⚙️ How Our Code Works

1. **Setup Reactor:** We define the High-Energy Hamiltonian—this is the raw plasma state inside the reactor core.
2. **Ignite the Sun:** We prepare the Qubit in an extremely energetic excited state `|1>`.
3. **Time-Evolution (Wrapping):** We apply rotations mimicking the flow of time and the crushing magnetic fields attempting to 'wrap' the sun.
4. **Thermal Readout:** We measure the output to check if the artificial star stabilized perfectly or collapsed into absolute darkness.

```
import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile
from qiskit_aer import AerSimulator

# Set seed to 81 (Surah At-Takwir)
np.random.seed(81)

print("☀️ [PHASE 1] Igniting Tokamak Nuclear Fusion Reactor Core...")

# 1 Qubit represents the heavily concentrated Solar Plasma
plasma_qr = QuantumRegister(1, 'plasma')
cr = ClassicalRegister(1, 'thermal_scan')
qc = QuantumCircuit(plasma_qr, cr)

# Step 1: Ignite the artificial sun (Placing it into high-energy state |1>)
qc.x(plasma_qr[0])
qc.barrier()
print("🔥 [PHASE 2] Plasma Generation Successful. Extreme temperatures reached.")

# Step 2: Time-Evolution (Trotterization)
# We apply a rotational gate representing a massive magnetic field wrapping around the plasma
# to prevent it from dropping into the 'ocean' and causing a sea of flames.
# We set an exact angle to stabilize the state without wrapping it in total darkness.
magnetic_containment = 0.44 
qc.rx(magnetic_containment, plasma_qr[0])
qc.barrier()
print("🧲 [PHASE 3] Time-Evolution applied. Magnetic fields wrapping the sun.")

# Step 3: Measurement (Thermal Check)
qc.measure(plasma_qr, cr)

simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1024)
counts = job.result().get_counts()

# We analyze if the state collapsed to 0 (Wrapped in darkness) or held its energy (1).
energy_retained = counts.get('1', 0)
darkness = counts.get('0', 0)

print("\n================================================")
print("          ⚛️ TOKAMAK FUSION CORE DIAGNOSTICS      ")
print("================================================")
print(f"☀️ Plasma Energy Successfully Maintained: {energy_retained} shots")
print(f"🌑 Reactor Extinguished (Wrapped in Darkness): {darkness} shots")
containment_stability = (energy_retained / 1024) * 100
print(f"✅ Artificial Sun Containment Stability: {containment_stability:.2f}%")
print("================================================")
```
