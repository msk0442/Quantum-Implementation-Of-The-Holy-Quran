### ⚛️ Code Pattern Mechanics: Quantum Phase Estimation (QPE)

*This project drops simple superposition loops in favor of advanced Fourier mathematics representing telescopy.*
1.  **Allocate the Sensor Array:** We create an estimation register (3 qubits representing the telescope's resolution) and a target register (1 qubit representing the incoming star's signal).
2.  **The Piercing Signal Input:** We encode an unknown, highly specific rotational phase into the target qubit, representing the hidden frequency of the distant Pulsar.
3.  **Controlled Filtering:** We apply series of controlled phase-rotations linking the star to our telescope's estimation array.
4.  **Inverse QFT (The Trial):** The algorithm applies the Inverse Quantum Fourier Transform. This is the ultimate "Trial" that rips away the noise, forcing the qubits to output the exact mathematical probability wave of the hidden star's pulsing frequency.

```
import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile
from qiskit.circuit.library import QFT
from qiskit_aer import AerSimulator

# Setting the physics simulation seed exactly to 86 (Surah At-Tariq)
np.random.seed(86)

print("🔭 [ASTROPHYSICS SENSOR] Booting Quantum Deep-Space Telemetry Array...")
print("✨ [TARGET ACQUIRED] Scanning for anomalous 'Piercing Star' Pulsar sequences in Sector 86.")

# -------------------------------------------------------------
# STEP 1: Telescope Setup
# -------------------------------------------------------------
# 3 Qubits for the Measurement/Resolution array
meas_qr = QuantumRegister(3, 'telescope_resolution')
# 1 Qubit representing the incoming cosmic signal
target_qr = QuantumRegister(1, 'pulsar_signal')
cr = ClassicalRegister(3, 'signal_extraction')

qc = QuantumCircuit(meas_qr, target_qr, cr)

# Initialize the Telescope array in superposition (scaning everywhere)
qc.h(meas_qr)

# Initialize the Target Signal into the |1> state to prepare for the phase encoding
qc.x(target_qr)
qc.barrier()

# -------------------------------------------------------------
# STEP 2: Encoding the Hidden "Piercing" Pulsar Frequency
# -------------------------------------------------------------
# The Pulsar is rotating/pulsing at a highly specific hidden frequency (Phase = 0.5 pi).
# We simulate the star piercing through the dark by applying this phase rotation to the target.
hidden_pulsar_phase = 0.5 * np.pi 

# The Telescope attempts to lock onto the signal using controlled phase gates 
# (Controlled - P Gates) corresponding to the resolution registers.
repetitions = 1
for counting_qubit in range(3):
    for i in range(repetitions):
        qc.cp(hidden_pulsar_phase, meas_qr[counting_qubit], target_qr[0])
    repetitions *= 2

qc.barrier()
print("📡 [SYNCING PHASE] Intercepting cosmic jets... Background noise highly saturated.")

# -------------------------------------------------------------
# STEP 3: The Trial of Secrets (Inverse QFT)
# -------------------------------------------------------------
# We apply the Inverse Quantum Fourier Transform.
# This forces the chaotic noise to collapse, perfectly ripping the hidden 
# phase frequency out of the darkness and exposing it to measurement.
qc.append(QFT(3, inverse=True), meas_qr)
qc.barrier()

# -------------------------------------------------------------
# STEP 4: Measurement Extraction
# -------------------------------------------------------------
qc.measure(meas_qr, cr)

print("🔍 [THE TRIAL] Executing Inverse-QFT... Exposing hidden phase telemetry.")

# Run the Array
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1024)
counts = job.result().get_counts()

print("\n=======================================================")
print("         🌟 PULSAR TELEMETRY EXTRACTION REPORT ")
print("=======================================================")

# Extracting the dominant, exposed signal frequency
exposed_signal = max(counts, key=counts.get)
confidence = (counts[exposed_signal] / 1024) * 100

print(f"🚨 TARGET VERIFIED: 'Piercing Star' successfully detected amidst deep space noise.")
print(f"✅ Extracted Baseline Frequency Signature: |{exposed_signal}>")
print(f"✅ Signal-to-Noise Confidence Ratio: {confidence:.1f}%")

if exposed_signal == '010': # Mathematical state of 0.5 Pi Phase isolated on 3 qubits
    print("STATUS: Perfect Resonance Lock. Hidden cosmic structures successfully exposed.")
print("=======================================================\n")
```
